```python
#Create a function to get information from 
def find_info(tag_name, class_name):
    li = []
    for item in soup.find_all(tag_name, {"class": class_name}):
        if item.get_text(strip=True) == 'NoneType':
            li.append("NaN")
        else:
            li.append(item.get_text(strip=True))
    return li
```

```python
urls =[]

for a in soup.find_all('a', href=True):
    if "/jobs/view/" in a['href']:
        urls.append(a['href'])

print(len(urls))
```

```python
"""The State attribute should be 
â€œEarly Applicationsâ€ if there is an alert that shows Early Applicant, 
"On-going" if there is an alert that shows Actively Hiring/Recruiting, 
or â€œOthersâ€ for alerts that are different than the two previously mentioned ones."""

state_li = []

for ji in soup.find_all("span", {"class": "result-benefits__text"}):
    if ji.get_text(strip=True) == 'NoneType':
        state_li.append("NaN")
    elif "early applicant" in ji.get_text(strip=True):
        state_li.append("Early Applications")
    elif "Actively Hiring" in ji.get_text(strip=True):
        state_li.append("On-going")
    else:
        state_li.append("Others")
        
len(state_li)
```

```python
original_titles = []
materials = []
dimensions = []
credits = []

for url in urls:
    
    source = requests.get(url)
    soup = bs4.BeautifulSoup(source.text, 'html.parser')
    
     # get details
    details = soup.find_all("div", {"class": "aside-detail"})
    
    for item in details:
        if "Original title" in item.text:
            original_titles.append(item.text.split("Original title")[1])
        elif "Medium/Materials" in item.text:
            materials.append(item.text.split("Medium/Materials")[1])
        elif "Dimensions" in item.text:
            dimensions.append(item.text.split("Dimensions")[1])
        elif "Credit line" in item.text:
            credits.append(item.text.split("Credit line")[1])

    original_titles.append(details[0].text)
    materials.append(details[2].text.split('\n')[0])
    dimensions.append(details[3].text)
    credits.append(details[4].text)
```
